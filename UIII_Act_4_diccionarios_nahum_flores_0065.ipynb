#### Diccionarios

```python
# 1. Inicio con un diccionario vacío
diccionario_disenador = {}

# --- FUNCIONES DEL SISTEMA ---

def listar_disenador():
    # MegaNota: La información se muestra en forma de tabla distribuida
    print("\n" + "="*85)
    print(f"{'ID':<10} | {'NOMBRE':<25} | {'ESPECIALIDAD':<20} | {'SOFTWARE':<15}")
    print("-" * 85)

    # Uso de for clave, valor (clave valor en for) como se solicitó
    for clave, valor in diccionario_disenador.items():
        print(f"{clave:<10} | {valor['Nombre']:<25} | {valor['Especialidad']:<20} | {valor['Software']:<15}")

    if not diccionario_disenador:
        print(f"{' ':>20}*** El diccionario está vacío actualmente ***")
    print("="*85)

def agregar_disenador():
    print("\n--- REGISTRAR NUEVO DISEÑADOR ---")
    id_dis = input("Ingrese ID único: ")
    nom = input("Ingrese Nombre Completo: ")
    esp = input("Ingrese Especialidad (Web, Gráfico, UI/UX): ")
    soft = input("Ingrese Software principal (Figma, Adobe, etc): ")

    # Agregar al diccionario
    diccionario_disenador[id_dis] = {
        "Nombre": nom,
        "Especialidad": esp,
        "Software": soft
    }
    print("\n>>> ¡Diseñador agregado con éxito!")
    listar_disenador() # Siempre utiliza la opción 2 después de cada acción

def buscar_disenador():
    id_dis = input("Ingrese el ID del diseñador a buscar: ")
    if id_dis in diccionario_disenador:
        datos = diccionario_disenador[id_dis]
        print("\n--- RESULTADO DE LA BÚSQUEDA ---")
        print(f"ID: {id_dis} | Nombre: {datos['Nombre']} | Especialidad: {datos['Especialidad']} | Software: {datos['Software']}")
    else:
        print("\n[!] Error: El ID no existe.")
    listar_disenador()

def borrar_disenador():
    id_dis = input("Ingrese el ID del diseñador a borrar: ")
    if id_dis in diccionario_disenador:
        del diccionario_disenador[id_dis]
        print(f"\n>>> El registro con ID {id_dis} ha sido eliminado.")
    else:
        print("\n[!] No se encontró ningún diseñador con ese ID.")
    listar_disenador()

def actualizar_disenador():
    id_dis = input("Ingrese el ID del diseñador que desea actualizar: ")
    if id_dis in diccionario_disenador:
        print("--- Ingrese los nuevos datos ---")
        nom = input("Nuevo Nombre: ")
        esp = input("Nueva Especialidad: ")
        soft = input("Nuevo Software: ")

        # Actualización de datos
        diccionario_disenador[id_dis] = {
            "Nombre": nom,
            "Especialidad": esp,
            "Software": soft
        }
        print("\n>>> Registro actualizado correctamente.")
    else:
        print("\n[!] El ID proporcionado no existe.")
    listar_disenador()

# --- MENÚ INTERACTIVO (While True e If/Elif) ---

while True:
    print("\n*********** MENÚ DE CONTROL: DISEÑADORES ***********")
    print("1. Agregar Diseñador")
    print("2. Listar Diseñadores")
    print("3. Buscar Diseñador")
    print("4. Borrar Diseñador")
    print("5. Actualizar Diseñador")
    print("6. Salir")

    opcion = input("Seleccione una opción (1-6): ")

    if opcion == "1":
        agregar_disenador()
    elif opcion == "2":
        listar_disenador()
    elif opcion == "3":
        buscar_disenador()
    elif opcion == "4":
        borrar_disenador()
    elif opcion == "5":
        actualizar_disenador()
    elif opcion == "6":
        print("Saliendo del programa... ¡Éxito en tu revisión!")
        break
    else:
        print("Opción no válida, por favor intenta de nuevo.")

```
